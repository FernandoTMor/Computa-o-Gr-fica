```python
import numpy as np
import matplotlib.pyplot as plt

# Definição dos pontos e vetores tangentes
P1 = np.array([0, 0])
P2 = np.array([3, 3])
M1 = np.array([6, -6])
M2 = np.array([6, 4])

# Função para calcular a curva de Hermite
def hermite_curve(t, P1, P2, M1, M2):
    H1 = 2*t**3 - 3*t**2 + 1
    H2 = t**3 - 2*t**2 + t
    H3 = -2*t**3 + 3*t**2
    H4 = t**3 - t**2
    return H1*P1 + H2*M1 + H3*P2 + H4*M2

# Criando os pontos para a curva com 10 segmentos
t_values = np.linspace(0, 1, 11)  # 10 segmentos → 11 pontos
curve_points = np.array([hermite_curve(t, P1, P2, M1, M2) for t in t_values])

# Plotando a curva
plt.figure(figsize=(8, 6))
plt.plot(curve_points[:, 0], curve_points[:, 1], 'purple', linewidth=2, label="Curva de Hermite")
plt.scatter(curve_points[:, 0], curve_points[:, 1], color="black", marker='s', label="Pontos da curva", zorder=3)
plt.scatter([P1[0], P2[0]], [P1[1], P2[1]], color="red", s=100, label="Pontos P1 e P2", zorder=4)
plt.quiver(P1[0], P1[1], M1[0], M1[1], color="blue", angles='xy', scale_units='xy', scale=10, label="M1")
plt.quiver(P2[0], P2[1], M2[0], M2[1], color="green", angles='xy', scale_units='xy', scale=10, label="M2")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Curva de Hermite - 10 Segmentos")
plt.axhline(0, color='black', linewidth=0.5)  # Eixo X
plt.axvline(0, color='black', linewidth=0.5)  # Eixo Y
plt.grid(True)
plt.legend()
plt.show()

```
