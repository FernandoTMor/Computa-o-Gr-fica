```python
import numpy as np


A = np.array([[2,3,8],[6,0,4],[1,5,7]])

linhas = len(A)
colunas = len(A[0])

transposta = [[0 for _ in range(linhas)] for _ in range(colunas)]

for i in range(linhas):
    for j in range(colunas):
        transposta[j][i] = A[i][j]

for linha in transposta:
    print(linha)

```
