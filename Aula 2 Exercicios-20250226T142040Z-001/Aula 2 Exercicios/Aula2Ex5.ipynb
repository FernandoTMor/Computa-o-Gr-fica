```python
import numpy as np

A = np.array([[2,3,8],[6,0,4],[1,5,7]])
x = 3

linhas = len(A)
colunas = len(A[0])

resultado = [[0 for _ in range(colunas)] for _ in range(linhas)]

for i in range(linhas):
    for j in range(colunas):
        resultado[i][j] = A[i][j] * x

for linha in resultado:
    print(linha)
```
