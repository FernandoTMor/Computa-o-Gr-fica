```python
import numpy as np

A = np.array([[2,0,0], [0,1,0], [0,0,7]])

def eh_diagonal(A):
  linhas = np.shape(A)[0]
  colunas = np.shape(A)[1]
  if linhas != colunas:
    return False

  for i in range(linhas):
    for j in range(colunas):
      if i != j and A[i,j] != 0:
        return False
  return True

if eh_diagonal(A):
  print("A é diagonal")
else:
  print("A não é diagonal")

```
