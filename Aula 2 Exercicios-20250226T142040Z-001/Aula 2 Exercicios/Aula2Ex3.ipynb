```python
import numpy as np

A = np.array([[1,0,0],[0,1,0],[0,0,1]])

def eh_identidade(A):
  linhas = np.shape(A)[0]
  colunas = np.shape(A)[1]

  if linhas!=colunas:
    return False

  for i in range(linhas):
    for j in range(colunas):
      if(i==j and A[i][j]!=1) or (i!=j and A[i][j]!=0):
        return False
  return True

if eh_identidade(A):
  print("A é identidade")
else:
  print("A não é identidade")
```
