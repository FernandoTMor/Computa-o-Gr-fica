```python
import numpy as np
import matplotlib.pyplot as plt

# Função para calcular a translação dos pontos
def rotacao(pontos, angulo):
    pontos_rotacionados = []
    rad = np.radians(angulo)
    cos = np.cos(rad)
    sin = np.sin(rad)

    for ponto in pontos:
        x_r = ponto[0] * cos - ponto[1] * sin
        y_r = ponto[0] * sin + ponto[1] * cos
        pontos_rotacionados.append((x_r, y_r))

    return pontos_rotacionados

# Pontos originais
p1 = (-1, -1)
p2 = (-1, 1)
p3 = (1, 1)
p4 = (1, -1)

# angulo da rotação
angulo = 45

# Calcular a translação dos pontos
pontos_originais = [p1, p2, p3, p4, p1]  # Adiciona p1 novamente para fechar o triângulo
pontos_rotacionados = rotacao(pontos_originais, angulo)

# Plotar os pontos originais e os pontos transladados
plt.plot([p[0] for p in pontos_originais], [p[1] for p in pontos_originais], 'bo-', label='Pontos originais')
plt.plot([p[0] for p in pontos_rotacionados], [p[1] for p in pontos_rotacionados], 'ro-', label='Pontos rotacionados')

# Configurações do gráfico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Translação de pontos no plano cartesiano')
plt.grid(True)
plt.legend()

# Mostrar o gráfico
plt.show()
```
